```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Student\Downloads\autos_new.csv')
df['unrepaired_damage'].unique()

df.iloc[0]

#df.groupby(by=['unrepaired_damage'])['name'].count()
df = df[['name','dollar_price','registration_year','kilometer','brand','unrepaired_damage','model','vehicle_type']]

# Filtrujeme na hodnotu 'nein', protoze chceme pouze nebourana auta
# Zaroven nas zajimaji pouze hodnoty skoda ze sloupce brand
# Pouzivame logickou podminku A za pomoci znaku &, jednotlive podminky oddelujeme zavorkami
df = df[(df['unrepaired_damage']=='nein') ]
# Uz mame pouze nebourana auta, sloupecek mazeme
df.drop('unrepaired_damage',inplace=True,axis=1)
# Dataset je z roku 2016, od tohoto roku pocitame stari vozu
df['stari'] = 2016 - df['registration_year']
# Filtrujeme na vozi mladsi ci rovno 20 let
df = df[df['stari']<=20]
df
```

```python
# Za pomoci .copy() vytvarime fyzickou kopii dataframe df
df1 = df.copy()

# Pro kazdou jednu hodnotu z unikatnich  hodnot ve sloupci vehicle_type
for  i in df1['vehicle_type'].unique():
    _df = df1[df1['vehicle_type']== i ]
# Korelační matici s rozdělením podle vehicle_type
    # Dataframe omezujeme pouze na ciselne sloupce, ktere pak vstupuji do korelace
    _df = _df.select_dtypes(['number'])
    # do promenne si ukladame korelacni matici
    foo = _df.corr()
    # Korelacni matici vypisujeem do konzole
    print(f'{i}',foo,'\n')
```
